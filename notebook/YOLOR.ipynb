## Introduction

Object Detection is a important problem in computer vision. In this notebook we are going to do object detection in a custom dataset (historical Data) using YOLOR.

YOLOR is a state-of-the-art machine learning algorithm for object detection, different from YOLOv1-YOLOv5 due to the difference in authorship, architecture, and model infrastructure. YOLOR stands for “You Only Learn One Representation".



**References:**
- [You Only Learn One Representation: Unified Network for Multiple Tasks Paper](https://arxiv.org/abs/2105.04206)

- [YOLOR github repository](https://github.com/augmentedstartups/yolor)

# Run the notebook on custom dataset

In order to run the notebook on custom dataset you need to organize the dataset directory in the following way:

```
dataset
└─── original dataset
|   |
|   └─── images
|   |   |
|   |   └─── test_dir
|   |   └─── train_dir
|   |   └─── val_dir
|   |
|   |
|   └─── labels
|   |   |
|   |   └─── test_dir
|   |   └─── train_dir
|   |   └─── val_dir

```

Under images each directory contains all the images used as training, validation and test sets.
Under labels each directory contains the labels for each image. \\
The image and its labels must have the same name.

The following requirements must also be met:
- take the new version of *plots.py* from [here](https://github.com/WongKinYiu/yolor/issues/113), and move it in the right position
- create a proper configuration file from your custom dataset. Information can be found [here](https://github.com/WongKinYiu/yolor/issues/113)

### Configuration file creation

Create configuration file and dataset setting

````python
#create custom.yaml file

# train and val datasets (image directory or *.txt file with image paths)
path="path: ../dataset/original_dataset"
train="train: dataset/original_dataset/images/train_dir"
val="val : dataset/original_dataset/images/val_dir"
test="test: dataset/original_dataset/images/test_dir"
# number of classes
nc="nc: 10"
# class names
names='names: ["fake","tank","car","truck","boat","airplane","horse trailer","train","motorcycle","Unknow"]'

#file position
custom_yaml_path="/content/yolor/data/custom.yaml"
#write everithing on the file 
f = open(custom_yaml_path , "w")
f.write("# train and val datasets (image directory or *.txt file with image paths)\n" 
        + path + "\n" + train + "\n" + val + "\n" + test + "\n\n" + 
        "# number of classes\n" + nc + "\n\n" + 
        "# class names\n" + names)
f.close()
````

````python
#create custom.names file
custom_names_path="/content/yolor/data/custom.names"
f = open(custom_names_path , "w")
names=["fake", "tank","car","truck","boat","airplane","horse trailer","train","motorcycle","Unknow"]
for c in names:
  f.write(c+"\n")
f.close()
````
